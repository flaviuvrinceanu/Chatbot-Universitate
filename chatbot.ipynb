```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# we use a Llama2 model from Hugging Face
model_name = "meta-llama/Llama-2-7b-chat-hf"

# load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

# pipeline creation
generator = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=128)
```

#Prompts

```python
prompturi = [
    "What are the main benefits of using chatbots in universities?",
    "Explain what is artificial intelligence in simple terms.",
    "Summarize the importance of data privacy in student services."
]

for i, prompt in enumerate(prompturi, 1):
    result = generator(prompt)
    print(f"\nPrompt {i}:\n{prompt}\nModel Response:\n{result[0]['generated_text']}\n{'-'*60}")
```
